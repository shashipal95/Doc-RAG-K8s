## 1. Document Processing
Logic for extracting text from various file formats.

```python
def extract_text(file_path):
    ext = os.path.splitext(file_path)[1].lower()
    if ext == '.pdf':
        reader = PdfReader(file_path)
        return "\n".join([page.extract_text() for page in reader.pages])
    elif ext == '.docx':
        doc = Document(file_path)
        return "\n".join([p.text for p in doc.paragraphs])
    return ""
```

## 2. Hybrid Search Setup
Implementing BM25 sparse vectors combined with dense embeddings.

```python
# Initialize Pinecone and BM25
pc = Pinecone(api_key=os.getenv("PINECONE_API_KEY"))
index = pc.Index(os.getenv("PINECONE_INDEX_NAME"))
bm25 = BM25Encoder().default()

def get_hybrid_vectors(text, gemini_client):
    # Dense vector
    dense = gemini_client.models.embed_content(
        model="text-embedding-004",
        parts=[types.Part.from_text(text)]
    ).embeddings[0].values
    
    # Sparse vector
    sparse = bm25.encode_queries(text)
    
    return dense, sparse
```
